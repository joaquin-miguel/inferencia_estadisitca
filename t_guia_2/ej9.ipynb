Modelo no param√©trico: comparaci√≥n de dos medias poblacionales. Usar el dataset del ejercicio 14 de la gu√≠a anterior de producci√≥n de lluvia en dos grupos distintos de nubes. 
Sea ùúá1 la precipitaci√≥n media generada por el grupo de nubes sin sembrar y ùúá2 la mediana del grupo de las nubes sembradas, 
realizar un test de nivel 0.05 para decidir si ùúá1 = ùúá2 (ùêª0) o ùúá1 ‚â† ùúá2 (ùêª1).


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = pd.read_csv('../t_guia_1/cloud-seeding.txt', sep='\t')
data.head()
```

```python
# Crear la figura y los ejes para los gr√°ficos
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Graficar el histograma y KDE para 'Unseeded Clouds'
axes[0].hist(data['Unseeded Clouds'], bins=10, color='skyblue', alpha=0.7, density=True, edgecolor='black')
axes[0].set_title('Unseeded Clouds')
axes[0].set_xlabel('Rainfall')
axes[0].set_ylabel('Density')

# Graficar el histograma y KDE para 'Seeded Clouds'
axes[1].hist(data['Seeded Clouds'], bins=10, color='orange', alpha=0.7, density=True, edgecolor='black')
axes[1].set_title('Seeded Clouds')
axes[1].set_xlabel('Rainfall')
axes[1].set_ylabel('Density')

# Ajustar y mostrar
plt.tight_layout()
plt.show()
```

Realizar un test de nivel 0.05 para decidir si 

ùêª0: ùúá1 = ùúá2  
ùêª1: ùúá1 ‚â† ùúá2

```python
# Sea ùúá1 la precipitaci√≥n media generada por el grupo de nubes sin sembrar y ùúá2 la mediana del grupo de las nubes sembradas
nubes_sin_sembrar = data['Unseeded Clouds']
nubes_sembradas = data['Seeded Clouds']

u_1 = np.mean(nubes_sin_sembrar)
u_2 = np.mean(nubes_sembradas)

print(f"Media de nubes sin sembrar: {u_1}")
print(f"Media de nubes sembradas: {u_2}")

print(len(nubes_sin_sembrar))
print(len(nubes_sembradas))
```
